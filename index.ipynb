Carbon dioxide (CO₂) has long been recognized as the dominant driver of climate change, and understanding how it's emitted by whom, from where, and in what quantity is critical for designing sustainable global policies. The dataset from Our World in Data serves as a powerful lens into this issue, allowing us to explore how CO₂ emissions have evolved across countries and centuries, and how they're intertwined with economic development and energy use.

Spanning from 1750 to 2023, this dataset presents a detailed historical record of emissions from nearly every country in the world. ## Change in Per Capita CO₂ Emissions and GDP

For Reducing the effect of size of population in different countries, We used Per Capita CO2 emmision as our parameter for evaluation


#### Developed Countries 

```python
def get_variation(values: pd.Series) -> np.float64:
    base = values.iloc[0]  
    current = values.iloc[-1]  

    return (current - base) / base if base else 0  

def get_df_changes(frame):

    variations = frame['gdp_per_capita'].expanding(min_periods=2).apply(get_variation)
    frame = frame.assign(gdp_per_capita=variations*100)

    variations = frame['co2_per_capita'].expanding(min_periods=2).apply(get_variation)
    frame = frame.assign(co2_per_capita=variations*100)

    variations = frame['consumption_co2_per_capita'].expanding(min_periods=2).apply(get_variation)
    frame = frame.assign(consumption_co2_per_capita=variations*100)

    frame = frame.set_index('year')
    frame = frame.fillna(0)

    return frame

newnames = {'gdp_per_capita':'GDP Per Capita', 'co2_per_capita': 'Production CO2 Per Capita', 'consumption_co2_per_capita': 'Consumption CO2 Per Capita'}
db['gdp_per_capita'] = db['gdp']/db['population']
economic_db = db[['gdp_per_capita', 'co2_per_capita', 'consumption_co2_per_capita', 'country', 'year']]

uk_df = economic_db.loc[economic_db['country'] == 'United Kingdom']
uk_df = uk_df.loc[economic_db['year'] >= 2000]
uk_df = uk_df.dropna()

uk_df = get_df_changes(uk_df)

fig = px.line(uk_df, x = uk_df.index, y = ['gdp_per_capita', 'co2_per_capita', 'consumption_co2_per_capita'], template="plotly_dark", markers=True, title='Change in Per Capita CO2 Emissions and GDP, United Kingdom')
fig.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                      legendgroup = newnames[t.name],
                                      hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                                     )
                  )
fig.update_layout( yaxis_title = 'Percentage Change with Base Year 2000', xaxis_title = 'Year', legend_title='')
fig.show()
```

```python
j_df = economic_db.loc[economic_db['country'] == 'Japan']
j_df = j_df.loc[economic_db['year'] >= 2000]
j_df = j_df.dropna()

j_df = get_df_changes(j_df)

fig = px.line(j_df, x = j_df.index, y = ['gdp_per_capita', 'co2_per_capita', 'consumption_co2_per_capita'], template="plotly_dark", markers=True, title='Change in Per Capita CO2 Emissions and GDP, Japan')
fig.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                      legendgroup = newnames[t.name],
                                      hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                                     )
                  )
fig.update_layout( yaxis_title = 'Percentage Change with Base Year 2000', xaxis_title = 'Year', legend_title='')
fig.show()
```

```python
n_df = economic_db.loc[economic_db['country'] == 'Norway']
n_df = n_df.loc[economic_db['year'] >= 2000]
n_df = n_df.dropna()

n_df = get_df_changes(n_df)

fig = px.line(n_df, x = n_df.index, y = ['gdp_per_capita', 'co2_per_capita', 'consumption_co2_per_capita'], template="plotly_dark", markers=True, title='Change in Per Capita CO2 Emissions and GDP, Norway')
fig.for_each_trace(lambda t: t.update(name = newnames[t.name],
                                      legendgroup = newnames[t.name],
                                      hovertemplate = t.hovertemplate.replace(t.name, newnames[t.name])
                                     )
                  )
fig.update_layout( yaxis_title = 'Percentage Change with Base Year 2000', xaxis_title = 'Year', legend_title='')
fig.show()
```

Seeing in the above plots that GDP and Co2-emmision have relation with each other, and their relation can define a country's properties, so We constructed a new parameter which is **Corelation between CO2 and GDP**.

## Comparing GDP vs corelation of GDP with CO₂ emissions

```python
correlation_data = []

for country in db['country'].unique():
    df_country = db[db['country'] == country]
    if df_country['gdp'].notna().sum() > 10 and df_country['co2'].notna().sum() > 10:
        df_clean = df_country[['year', 'gdp', 'co2']].dropna()
        if len(df_clean) > 10:
            correlation = df_clean['gdp'].corr(df_clean['co2'])
            latest_gdp = df_clean.sort_values('year')['gdp'].iloc[-1]
            correlation_data.append({
                'country': country,
                'gdp': latest_gdp,
                'correlation': correlation
            })

corr_df = pd.DataFrame(correlation_data)
corr_df = corr_df.replace([np.inf, -np.inf], np.nan).dropna()

fig = px.scatter(
    corr_df,
    x='gdp',
    y='correlation',
    hover_name='country',
    log_x=True,
    labels={
        'gdp': 'GDP (log scale)',
        'correlation': 'Correlation (GDP vs CO₂ Emissions)'
    },
    title='GDP vs Correlation of GDP with CO₂ Emissions by Country',
    template='plotly_dark',
    color='correlation',
    color_continuous_scale='Viridis'
)

fig.update_traces(marker=dict(size=8, line=dict(width=1, color='DarkSlateGrey')))
fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import dash
from dash import dcc, html, Input, Output

owid = pd.read_csv("https://raw.githubusercontent.com/Rajput-Vivek-16/CO2-Emission-Analysis/main/_notebooks/data/owid-co2-data.csv")
owid = owid[owid['iso_code'].str.len() == 3]  

events = {
    "Kyoto Protocol (1997)": (1995, 2000),
    "Paris Agreement (2015)": (2013, 2018),
    "COVID-19 Pandemic (2020)": (2018, 2021)
}

db = owid[['country', 'year', 'gdp', 'co2', 'co2_per_capita', 'energy_per_capita']].dropna(subset=['gdp', 'co2'])
db = db[~db['country'].str.contains('World|Africa|Europe|Asia|Oceania|International')]
countries = sorted(db['country'].unique())

app = dash.Dash(__name__, suppress_callback_exceptions=True)
app.title = "CO₂ Dashboard"

dropdown_style = {
    'color': 'black',
    'borderRadius': '10px',
    'fontFamily': 'Poppins, sans-serif',
    'height': '36px',
    'fontSize': '14px'
}

tab_style = {
    'padding': '10px',
    'borderRadius': '10px',
    'backgroundColor': '#333',
    'color': '#fff',
    'fontFamily': 'Poppins, sans-serif',
    'marginRight': '5px'
}

selected_tab_style = {
    'padding': '10px',
    'borderRadius': '10px',
    'backgroundColor': '#555',
    'color': '#fff',
    'fontWeight': 'bold',
    'fontFamily': 'Poppins, sans-serif',
    'marginRight': '5px'
}

app.layout = html.Div(style={'backgroundColor': '#111', 'color': 'white', 'padding': '10px', 'fontFamily': 'Poppins, sans-serif'}, children=[
    html.H1("CO₂ Emissions Dashboard", style={'textAlign': 'center'}),

    dcc.Tabs(id="tabs", value='tab-events', children=[
        dcc.Tab(label='Global Events View', value='tab-events', style=tab_style, selected_style=selected_tab_style),
        dcc.Tab(label='Country Comparison View', value='tab-countries', style=tab_style, selected_style=selected_tab_style),
        dcc.Tab(label='What-If Simulator', value='tab-whatif', style=tab_style, selected_style=selected_tab_style)
    ]),

    html.Div(id='tabs-content')
])

@app.callback(Output('tabs-content', 'children'), Input('tabs', 'value'))
def render_content(tab):
    if tab == 'tab-events':
        return html.Div([
            html.Div([
                html.Label("Select Event:", style={'fontFamily': 'Poppins, sans-serif'}),
                dcc.Dropdown(
                    id='event-dropdown',
                    options=[{"label": k, "value": k} for k in events.keys()],
                    value="COVID-19 Pandemic (2020)",
                    style=dropdown_style
                )
            ], style={'width': '40%', 'display': 'inline-block', 'padding': '10px 10px 10px 0'}),

            html.Div([
                html.Label("Select Number of Top Countries:", style={'fontFamily': 'Poppins, sans-serif'}),
                dcc.Dropdown(
                    id='top-n-dropdown',
                    options=[{"label": f"Top {n}", "value": n} for n in [10, 20, 30, 50, 100]],
                    value=10,
                    style=dropdown_style
                )
            ], style={'width': '40%', 'display': 'inline-block', 'padding': '10px 0 10px 10px'}),

            dcc.Graph(id='dumbbell-plot')
        ])

    elif tab == 'tab-countries':
        return html.Div([
            html.Div([
                html.Label("Select Country 1", style={'fontFamily': 'Poppins, sans-serif'}),
                dcc.Dropdown(countries, value='India', id='country1', style=dropdown_style),

                html.Label("Select Country 2", style={'marginTop': '20px', 'fontFamily': 'Poppins, sans-serif'}),
                dcc.Dropdown(countries, value='United States', id='country2', style=dropdown_style)
            ], style={'width': '40%', 'display': 'inline-block', 'padding': '10px'}),

            html.Div([
                dcc.Graph(id='gdp_co2_plot'),
                dcc.Graph(id='efficiency_plot'),
                dcc.Graph(id='co2_per_capita_plot')
            ])
        ])

    elif tab == 'tab-whatif':
        return html.Div([
            html.Div([
                html.Label("Select Country", style={'fontFamily': 'Poppins, sans-serif'}),
                dcc.Dropdown(countries, value='India', id='whatif-country', style=dropdown_style),

                html.Label("Select Scenario", style={'marginTop': '20px', 'fontFamily': 'Poppins, sans-serif'}),
                dcc.Dropdown(
                    id='scenario-type',
                    options=[
                        {'label': ' CO₂ Emissions', 'value': 'reduction'},
                        {'label': 'CO₂ Efficiency', 'value': 'efficiency'}
                    ],
                    value='reduction',
                    style=dropdown_style
                ),

                html.Label("Reduction/Increase Rate (% per year)", style={'marginTop': '20px', 'fontFamily': 'Poppins, sans-serif'}),
                dcc.Input(
                    id='reduction-rate',
                    type='number',
                    value=-10,
                    min=-100,
                    max=100,
                    step=0.1,
                    style={'width': '100%', 'marginBottom': '20px', 'padding': '8px', 'borderRadius': '8px'}
                ),

                html.Label("Target Year", style={'marginTop': '20px', 'fontFamily': 'Poppins, sans-serif'}),
                dcc.Input(id='target-year', type='number', value=2035, min=2025, max=2100, step=1,
                          style={'width': '100%', 'marginBottom': '20px'})
            ], style={'width': '40%', 'display': 'inline-block', 'padding': '10px'}),

            html.Div([
                dcc.Graph(id='whatif-plot')
            ])
        ])

@app.callback(
    Output('dumbbell-plot', 'figure'),
    Input('event-dropdown', 'value'),
    Input('top-n-dropdown', 'value')
)
def update_dumbbell_plot(event_name, top_n):
    year1, year2 = events[event_name]
    df1 = owid[owid['year'] == year1][['country', 'co2']].rename(columns={'co2': 'co2_before'})
    df2 = owid[owid['year'] == year2][['country', 'co2']].rename(columns={'co2': 'co2_after'})

    df = pd.merge(df1, df2, on='country').dropna()
    df['abs_change'] = (df['co2_after'] - df['co2_before']).abs()
    df['change'] = df['co2_after'] - df['co2_before']

    df = df.sort_values('abs_change', ascending=False).head(top_n)
    df = df.sort_values('abs_change', ascending=True)

    x_min = min(df['co2_before'].min(), df['co2_after'].min()) * 0.9
    x_max = max(df['co2_before'].max(), df['co2_after'].max()) * 1.1

    fig = go.Figure()

    for _, row in df.iterrows():
        fig.add_trace(go.Scatter(
            x=[row['co2_before'], row['co2_after']],
            y=[row['country'], row['country']],
            mode='lines',
            line=dict(color='gray', width=2),
            showlegend=False
        ))

    fig.add_trace(go.Scatter(
        x=df['co2_before'],
        y=df['country'],
        mode='markers',
        marker=dict(color='crimson', size=8),
        name=f'{year1}'
    ))

    fig.add_trace(go.Scatter(
        x=df['co2_after'],
        y=df['country'],
        mode='markers',
        marker=dict(color='limegreen', size=8),
        name=f'{year2}'
    ))

    fig.update_layout(
        title=f"Top {top_n} CO₂ Emission Changes (Absolute): {event_name}",
        xaxis_title="CO₂ Emissions (Million Tonnes)",
        yaxis_title="Country",
        xaxis=dict(range=[x_min, x_max]),
        yaxis=dict(type='category'),
        height=800,
        template='plotly_dark',
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=-0.3,
            xanchor="center",
            x=0.5
        ),
        margin=dict(l=100, r=20, t=80, b=80)
    )
    return fig

@app.callback(
    Output('gdp_co2_plot', 'figure'),
    Output('efficiency_plot', 'figure'),
    Output('co2_per_capita_plot', 'figure'),
    Input('country1', 'value'),
    Input('country2', 'value')
)
def update_country_graphs(country1, country2):
    df1 = db[db['country'] == country1]
    df2 = db[db['country'] == country2]

    fig1 = go.Figure()
    fig1.add_trace(go.Scatter(x=df1['gdp'], y=df1['co2'], mode='lines+markers', name=country1))
    fig1.add_trace(go.Scatter(x=df2['gdp'], y=df2['co2'], mode='lines+markers', name=country2))
    fig1.update_layout(title='GDP vs CO₂ Emissions', xaxis_title='GDP', yaxis_title='CO₂ Emissions', template='plotly_dark')

    df1['efficiency'] = df1['gdp'] / df1['co2']
    df2['efficiency'] = df2['gdp'] / df2['co2']
    fig2 = go.Figure()
    fig2.add_trace(go.Scatter(x=df1['year'], y=df1['efficiency'], mode='lines+markers', name=country1))
    fig2.add_trace(go.Scatter(x=df2['year'], y=df2['efficiency'], mode='lines+markers', name=country2))
    fig2.update_layout(title='CO₂ Efficiency (GDP per unit CO₂)', xaxis_title='Year', yaxis_title='Efficiency', template='plotly_dark')

    fig3 = go.Figure()
    fig3.add_trace(go.Scatter(x=df1['year'], y=df1['co2_per_capita'], mode='lines+markers', name=country1))
    fig3.add_trace(go.Scatter(x=df2['year'], y=df2['co2_per_capita'], mode='lines+markers', name=country2))
    fig3.update_layout(title='CO₂ Emissions per Capita', xaxis_title='Year', yaxis_title='CO₂ per Capita (tons)', template='plotly_dark')

    return fig1, fig2, fig3

@app.callback(
    Output('whatif-plot', 'figure'),
    Input('whatif-country', 'value'),
    Input('scenario-type', 'value'),
    Input('reduction-rate', 'value'),
    Input('target-year', 'value')
)
def update_whatif_plot(country, scenario, rate, target_year):
    # Handle empty inputs by providing default values or skipping calculations
    if not country or not scenario or rate is None or not target_year:
        # Return a default plot (e.g., empty plot or placeholder)
        return go.Figure(
            data=[],
            layout={
                'title': 'Please provide all inputs to generate the plot.',
                'xaxis': {'title': 'Year'},
                'yaxis': {'title': 'CO₂ Emissions (Million Tonnes)'},
                'template': 'plotly_dark'
            }
        )

    # Ensure rate and target_year are valid numbers
    try:
        rate = float(rate)
        target_year = int(target_year)
    except ValueError:
        return go.Figure(
            data=[],
            layout={
                'title': 'Invalid input values.',
                'xaxis': {'title': 'Year'},
                'yaxis': {'title': 'CO₂ Emissions (Million Tonnes)'},
                'template': 'plotly_dark'
            }
        )

    df = db[db['country'] == country]

    earliest_year = df['year'].min()
    historical_df = df[df['year'] >= earliest_year]

    current_year = historical_df['year'].max()

    years = list(range(current_year, target_year + 1))
    projected_values = []  # Store projected values (either CO₂ or efficiency)

    # Calculate the projected values for each year in the range
    for year in years:
        years_to_project = year - current_year
        if scenario == 'reduction':
            # For reduction, calculate projected CO₂ emissions
            current_co2 = historical_df[historical_df['year'] == current_year]['co2'].values[0]
            projected_value = current_co2 * (1 + rate / 100) ** years_to_project
        elif scenario == 'efficiency':
            # For efficiency, calculate projected efficiency (gdp / co2)
            current_gdp = historical_df[historical_df['year'] == current_year]['gdp'].values[0]
            current_co2 = historical_df[historical_df['year'] == current_year]['co2'].values[0]
            current_efficiency = current_gdp / current_co2  # Historical efficiency
            projected_efficiency = current_efficiency * (1 + rate / 100) ** years_to_project
            projected_value = projected_efficiency
        else:
            projected_value = None  # Default case (if no scenario selected)

        projected_values.append(projected_value)

    fig = go.Figure()

    if scenario == 'reduction':
        fig.add_trace(go.Scatter(
            x=historical_df['year'],
            y=historical_df['co2'],
            mode='lines+markers',
            name='Historical CO₂ emissions',
            line=dict(color='royalblue'),  # Change to royalblue
            marker=dict(size=6)
        ))
    elif scenario == 'efficiency':
        historical_df['efficiency'] = historical_df['gdp'] / historical_df['co2']  # Calculate efficiency for historical data
        fig.add_trace(go.Scatter(
            x=historical_df['year'],
            y=historical_df['efficiency'],
            mode='lines+markers',
            name='Historical Efficiency',
            line=dict(color='royalblue'),  # Change to royalblue
            marker=dict(size=6)
        ))

    fig.add_trace(go.Scatter(
        x=years,
        y=projected_values,
        mode='lines+markers',
        name=f'{country} - Projection',
        line=dict(color='crimson'),  # Change to crimson
        marker=dict(size=4, color='crimson')
    ))

    if scenario == 'reduction':
        fig.update_layout(
            title=f"{country} CO₂ Projection ({scenario})",
            xaxis_title="Year",
            yaxis_title="CO₂ Emissions (Million Tonnes)",
            template='plotly_dark',
            margin=dict(l=100, r=20, t=80, b=80)
        )
    elif scenario == 'efficiency':
        fig.update_layout(
            title=f"{country} Efficiency Projection ({scenario})",
            xaxis_title="Year",
            yaxis_title="Efficiency (GDP per unit CO₂)",
            template='plotly_dark',
            margin=dict(l=100, r=20, t=80, b=80)
        )

    return fig


if __name__ == '__main__':
    app.run(debug=True)
```

In this part we tried to diiferrentiate which countries are coal-based economies and which are oil based economies

```python


reliance_data = []
db = clean_df(raw_db)
db["iso_code"] = db["iso_code"].fillna(method="ffill").fillna(method="bfill")

for country in db['country'].unique():
    df_country = db[db['country'] == country]
    df_clean = df_country[['year', 'coal_co2_per_capita', 'oil_co2_per_capita', 'co2_per_capita']].dropna()
    if not df_clean.empty:
        latest_row = df_clean.sort_values('year').iloc[-1]
        reliance = 'Coal Economy' if latest_row['coal_co2_per_capita'] > latest_row['oil_co2_per_capita'] else 'Oil Economy'
        reliance_data.append({
            'country': country,
            'coal_co2_per_capita': latest_row['coal_co2_per_capita'],
            'oil_co2_per_capita': latest_row['oil_co2_per_capita'],
            'co2_per_capita': latest_row['co2_per_capita'],
            'reliance': reliance
        })

reliance_df = pd.DataFrame(reliance_data)

fig = px.scatter(
    reliance_df,
    x='coal_co2_per_capita',
    y='co2_per_capita',
    color='reliance',
    symbol='reliance',
    hover_name='country',
    labels={
        'coal_co2_per_capita': 'Coal CO₂ per Capita (tons per person)',
        'co2_per_capita': 'Total CO₂ per Capita (tons per person)'
    },
    title='Coal vs. Oil Economies and Their CO₂ Emissions per Capita',
    template='plotly_dark',
    color_discrete_sequence=['#4169E1', '#FF4500'],
    symbol_sequence=['circle', 'x'],
    height=700,
    width=1000
)

fig.update_traces(marker=dict(
    size=10,
    line=dict(width=0.5, color='white'),
    opacity=0.9
))

fig.update_layout(
    title_font_size=22,
    legend_title_text='Primary Fuel Reliance:',
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = db1[db1['year'] >= 1950][[
    'year', 
    'temperature_change_from_co2', 
    'temperature_change_from_ch4', 
    'temperature_change_from_n2o', 
    'co2', 
    'methane', 
    'nitrous_oxide'
]].dropna()

yearly = df.groupby('year').agg({
    'temperature_change_from_co2': 'mean',
    'temperature_change_from_ch4': 'mean',
    'temperature_change_from_n2o': 'mean',
    'co2': 'sum',
    'methane': 'sum',
    'nitrous_oxide': 'sum'
}).reset_index()

# Convert emissions from million tonnes to kg:
# 1 million tonnes = 1e6 tonnes, and 1 tonne = 1000 kg; hence multiply by 1e9.
yearly['co2_kg'] = yearly['co2'] * 1e9
yearly['methane_kg'] = yearly['methane'] * 1e9
yearly['nitrous_oxide_kg'] = yearly['nitrous_oxide'] * 1e9

# Compute the per kg effect on temperature change for each gas.
# (These values are expected to be very small.)
yearly['CO2_per_kg'] = yearly['temperature_change_from_co2'] / yearly['co2_kg']
yearly['CH4_per_kg'] = yearly['temperature_change_from_ch4'] / yearly['methane_kg']
yearly['N2O_per_kg'] = yearly['temperature_change_from_n2o'] / yearly['nitrous_oxide_kg']

melted = yearly.melt(id_vars='year', 
                     value_vars=['CO2_per_kg', 'CH4_per_kg', 'N2O_per_kg'],
                     var_name='Gas', value_name='Temperature Change per kg (°C)')

melted['Gas'] = melted['Gas'].replace({
    'CO2_per_kg': 'CO₂', 
    'CH4_per_kg': 'CH₄', 
    'N2O_per_kg': 'N₂O'
})

fig = px.line(melted, 
              x='year', 
              y='Temperature Change per kg (°C)', 
              color='Gas',
              labels={
                  'year': 'Year',
                  'Temperature Change per kg (°C)': 'Temperature Change per kg (°C)'
              },
              title='Per kg Effect of Different Gases on Temperature Change Over Time (1940–Present)',
              template='plotly_dark')

fig.update_layout(title_font_size=22, legend_title_font_size=14)
fig.show()
```

- Plotting the per-kg impact of different gases to see if Co2 is a major contributor in temperature change despite its large amount
- We can see that per-kg impact of methane is more than Co2

So, we can derive the order of most harmful greenhouse gases as methane > CO2 > N2O
